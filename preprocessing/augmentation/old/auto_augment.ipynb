```python
import imgaug as ia
import imgaug.augmenters as iaa
from imgaug.augmentables.bbs import BoundingBox, BoundingBoxesOnImage
import pandas as pd

# Load in Training Images
# Define Data Directory
data_dir = './train_imgs/'
data_test_lbls = pd.read_csv('./labelsTest.csv')
data_train_lbls = pd.read_csv('./labelsTrain.csv')
data_lbls = pd.concat([data_test_lbls, data_train_lbls])

def load_images(data_dir,resize=None):
    img_list = []
    img_lbl = []
    img_x1 = []
    img_x2 = []
    img_y1 = []
    img_y2 = []
    filenames = glob(data_dir + "*.jpg")
    filenames = [os.path.split(f)[1] for f in filenames]
    # Get Images
    for filename in filenames[:1]:
        print(filename)
        image = np.array(Image.open(data_dir+filename))
        img_list.append(image)
        
        img_desc = data_lbls.loc[data_lbls['filename'] == filename,:]
        
        lbl = img_desc['class'].values[0]
        x1 = img_desc['xmin'].values[0]
        x2 = img_desc['xmax'].values[0]
        y1 = img_desc['ymin'].values[0]
        y2 = img_desc['ymax'].values[0]
        print(lbl)
        print(x1)
        print(x2)
        print(x2)
        print(y2)
        
        img_lbl.append(lbl)
        img_x1.append(x1)
        img_x2.append(x2)
        img_y1.append(y1)
        img_y2.append(y2)
        
        # image = ia.quokka(size=(256, 256))
        bbs = BoundingBoxesOnImage([BoundingBox(x1=x1, y1=y1, x2=x2, y2=y2)], shape=image.shape)

        seq = iaa.Sequential([
            iaa.AdditiveGaussianNoise(scale=(0, 0.05*255))    
        ])

        # Augment BBs and images.
        image_aug, bbs_aug = seq(image=image, bounding_boxes=bbs)

        # print coordinates before/after augmentation (see below)
        # use .x1_int, .y_int, ... to get integer coordinates
        for i in range(len(bbs.bounding_boxes)):
            before = bbs.bounding_boxes[i]
            after = bbs_aug.bounding_boxes[i]
            print("BB %d: (%.4f, %.4f, %.4f, %.4f) -> (%.4f, %.4f, %.4f, %.4f)" % (
                i,
                before.x1, before.y1, before.x2, before.y2,
                after.x1, after.y1, after.x2, after.y2)
            )

        # image with BBs before/after augmentation (shown below)
        image_before = bbs.draw_on_image(image, size=2)
        image_after = bbs_aug.draw_on_image(image_aug, size=2, color=[0, 0, 255])
        
        plt.imshow(image_after)

    images = np.array(img_list)
    labels = []
    return images, img_lbl, img_x1, img_x2, img_y1, img_y2
  
# train_dir = './train_imgs/'
# train_imgs, train_labels, train_x1, train_x2, train_y1, train_y2 = load_images(train_dir)
```
